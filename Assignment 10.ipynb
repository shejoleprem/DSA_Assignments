class Solution {

    public ListNode removeElements(ListNode head, int val) {
        while(head!=null && head.val==val){
            head=head.next;
        }
        if(head==null){
            return null;
        }
        ListNode temp=head;
        while(temp.next!=null){
            ListNode dummy=temp.next;
            if(dummy.val==val){
                temp.next=temp.next.next;
            }else{
            temp=temp.next;
            }
        }
        return head;
    }
}

## Q4
public class Solution {

    public ListNode detectCycle(ListNode head) {
        HashSet<ListNode>h=new HashSet<>();
        ListNode n=head;
        while(n!=null){
            if(h.contains(n)){
                return n;
            }else{
                h.add(n);
            }
            n=n.next;
        }
        return null;
    }
}

## Q5
class Solution {

    public ListNode removeNthFromEnd(ListNode head, int n) {
        
        ListNode dummy=head;
        int size=0;
        while(dummy!=null){
            size++;
            dummy=dummy.next;
        }
        if(n==size){
            head=head.next;
            return head;
        }
        int index=size-n;
        int i=1;
        ListNode temp=head;
        while(i<index){
            temp=temp.next;
            i++;
        }
        temp.next=temp.next.next;

        return head;
    }
}

class Solution {

    public ListNode rotateRight(ListNode head, int k) {
        if(k<1 || head==null || head.next==null){
            return head;
        }
        int len=0;
        ListNode tail=head;
        while(tail!=null){
            tail=tail.next;
            len++;
        }
        k=k%len;
        for(int i=0;i<k;i++){
            head=Rotate(head);
        }

        return head;
        
    }

    public static ListNode Rotate(ListNode head){
        if(head.next==null){
            return head;
        }
        ListNode slow=head;
        ListNode fast=head.next;
        while(fast.next!=null){
            slow=slow.next;
            fast=fast.next;

        }
        slow.next=null;
        fast.next=head;
        head=fast;
        return head;
    }
}

## Q7
class Solution {

    public ListNode removeZeroSumSublists(ListNode head) {
       int sum=0;
       ListNode dm=new ListNode(0);
       dm.next=head;

       Map<Integer,ListNode>m=new HashMap<>();
        m.put(0,dm);

        for(ListNode i=dm;i!=null;i=i.next){
            sum+=i.val;
            m.put(sum,i);
        }

        sum=0;
        for(ListNode i=dm;i!=null;i=i.next){
            sum+=i.val;
            i.next=m.get(sum).next;
        }
        return dm.next;
    }
}
