package Assignment_6;

import java.util.Arrays;

public class Q1 {

    public static void main(String[] args) {
       String s = "IDID";
       int arr[]=diStringMatch(s);
        System.out.println(Arrays.toString(arr));
    }

    public static int[] diStringMatch(String s) {
        int hi=s.length();
        int lo=0;
        int k=0;
        int[] arr=new int[s.length()+1];
        for(int i=0;i<s.length();i++){
            if(s.charAt(i)=='I'){
                arr[k]=lo;
                lo++;
            }else if(s.charAt(i)=='D'){
                arr[k]=hi;
                hi--;
            }
            k++;
        }
        arr[s.length()]=lo;

        return arr;
    }
}

public class Q2 {

    public static void main(String[] args) {
       int[][] matrix = {{1,3,5,7},{10,11,16,20},{23,30,34,60}};
       boolean ans=searchMatrix(matrix,3);
        System.out.println(ans);
    }
    public static boolean searchMatrix(int[][] matrix, int target) {
        int rowlen=matrix.length;
        int collen=matrix[0].length;

        int low=0;
        int high=rowlen*collen-1;

        while(low<=high){
            int midindx=low+(high-low)/2;
            int rowindx=midindx / collen;
            int colindx=midindx % collen;

            int midelement=matrix[rowindx][colindx];

            if(midelement==target){
                return true;
            }
            else if(midelement<target){
                low=midindx+1;
            }
            else {
                high=midindx-1;
            }
        }
        return false;
    }
}

package Assignment_6;

import java.util.*;

public class Q4 {

    public static void main(String[] args) {
        int[]nums = {0,1};
        int ans=findMaxLength(nums);
        System.out.println(ans);
    }
    public static int findMaxLength(int[] nums) {
        int ans=0;
        HashMap<Integer,Integer>map=new HashMap<>();
        map.put(0,-1);
        int sum=0;

        for(int i=0;i<nums.length;i++){
            if(nums[i]==0){
                sum+=-1;
            }else if(nums[i]==1){
                sum+=1;
            }

            if(map.containsKey(sum)){
                int indx=map.get(sum);
                int len=i-indx;
                ans=Math.max(ans,len);
            }
            else{
                map.put(sum,i);
            }
        }
        return ans;
    }
}

package Assignment_6;

import java.util.Arrays;

public class Q5 {

    public static void main(String[] args) {
       int[] nums1 = {5,3,4,2};
       int[] nums2 = {4,2,2,5};
       int ans=minprodsum(nums1,nums2);
        System.out.println(ans);
    }
    public static int minprodsum(int[]nums1,int nums2[]){
        Arrays.sort(nums1);
        Arrays.sort(nums2);
        int res=0;
        for(int i=0;i<nums1.length;i++){
            res+=nums1[i]*nums2[nums2.length-i-1];
        }
        return res;
    }
}

package Assignment_6;

import java.util.*;

public class Q6 {

    public static void main(String[] args) {
        int[]changed = {1,3,4,2,6,8};
        int ans[]=findOriginalArray(changed);
        System.out.println(Arrays.toString(ans));
    }
    public static int[] findOriginalArray(int[] changed) {

        int len = changed.length;
        if((len&1) != 0) return new int[0];

        // Sorting the array
        Arrays.sort(changed);

        // Store frequencies in map
        Map<Integer,Integer> map = new HashMap<>();
        for(int e : changed) map.put(e,map.getOrDefault(e,0)+1);

        int[] res = new int[len/2];
        int k = 0;
        for(int i=0; i<len; i++){
            int ele = changed[i];

            // if map contains 'ele'
            if(map.containsKey(ele)){

                // if map contains 'ele*2'
                if(map.containsKey(ele*2)){
                    res[k++] = ele;

                    // reduce frequency of 'ele' and 'ele*2'
                    map.put(ele,map.get(ele)-1);
                    map.put(ele*2,map.get(ele*2)-1);

                    // if freq of any key becomes <=0, remove it from map
                    if(map.get(ele)<=0) map.remove(ele);
                    if(map.containsKey(ele*2) && map.get(ele*2)<=0) map.remove(ele*2);
                }
                else return new int[0];
            }

        }
        return res;
    }
}

package Assignment_6;

public class Q7 {

    public static void main(String[] args) {
        int ans[][]=generateMatrix(3);
        for(int i=0;i<3;i++){
            for(int j=0;j<3;j++){
                System.out.print(ans[i][j]+" ");
            }
            System.out.println();
        }
    }
    public static int[][] generateMatrix(int n) {
        int [][]arr=new int[n][n];
        int left=0;
        int top=0;
        int right=n-1;
        int bottom=n-1;
        int cnt=1;

        for(int j=0;j<(n+1)/2;j++){

            for(int i=left;i<=right;i++){
                arr[top][i]=cnt;
                cnt++;
            }
            top++;

            for(int i=top;i<=bottom;i++){
                arr[i][right]=cnt;
                cnt++;

            }
            right--;
            for(int i=right;i>=left;i--){
                arr[bottom][i]=cnt;
                cnt++;

            }
            bottom--;
            for(int i=bottom;i>=top;i--){
                arr[i][left]=cnt;
                cnt++;

            }
            left++;
        }

        return arr;

    }
}

package Assignment_6;

public class Q8 {

    public static void main(String[] args) {
        int[][]mat1 = {{1,0,0},{-1,0,3}};
        int[][]mat2 = {{7,0,0},{0,0,0},{0,0,1}};

        int ans[][]=multiplication(mat1,mat2);
        for(int i=0;i<ans.length;i++){
            for(int j=0;j<ans[0].length;j++){
                System.out.print(ans[i][j]+" ");
            }
            System.out.println();
        }
    }
    public static int[][] multiplication(int arr1[][],int arr2[][]){

        int ans[][]=new int[arr1.length][arr2[0].length];

        for(int i=0;i<ans.length;i++){
            for(int j=0;j<ans[0].length;j++){
                for(int k=0;k<arr2.length;k++){
                    ans[i][j]+=arr1[i][j]*arr2[k][j];
                }
            }
        }
        return ans;
    }
}
