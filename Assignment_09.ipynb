class Solution {

    public int[] twoSum(int[] numbers, int target) {
        int arr[]=new int[2];
        int s=0;
        int e=numbers.length-1;
        while(s<=e){
            int sum=numbers[s]+numbers[e];
            if(sum==target){
                arr[0]=s+1;
                arr[1]=e+1;
                return arr;
            }
            else if(sum>target){
                e--;
            }
            else{
                s++;
            }
        }
        return arr;
    }
}

class Solution {

    public int[] searchRange(int[] nums, int target) {
        int[]arr=new int[2];
        Arrays.fill(arr,-1);
        for(int i=0;i<nums.length;i++){
            if(nums[i]==target){
                arr[0]=i;
                break;
            }
        }
        for(int j=nums.length-1;j>=0;j--){
            if(nums[j]==target){
                arr[1]=j;
                break;
            }
        }
        return arr;
    }
}

class Solution {

    public int findPeakElement(int[] nums) {
        if(nums.length==1){
            return 0;
        }else if(nums.length==2){
            if(nums[0]>nums[1]){
                return 0;
            }else{
                return 1;
            }
        }
        int a=ans(nums,0,nums.length-1);
        return a;
    }
    public static int ans(int arr[],int s,int e){
        int mid=s+(e-s)/2;
        while(s<e){
            if(arr[mid+1]>arr[mid]){
                s=mid+1;
            }
            else{
                e=mid;
            }
            mid=s+(e-s)/2;
        }
        return s;
    }
}

class Solution {

    public int searchInsert(int[] nums, int target) {
        if(target<nums[0]){
            return 0;
        }else if(target>nums[nums.length-1]){
            return nums.length;
        }
        int si=0;
        int ei=nums.length-1;

        int mid=si+(ei-si)/2;

        while(si<=ei){
            if(nums[mid]==target){
                return mid;
            }else if(target>nums[mid]){
                si=mid+1;
            }else if(target<nums[mid]){
                ei=mid-1;
            }
            mid=si+(ei-si)/2;
        }
        return mid;
    }
    
}

class Solution {

    public int majorityElement(int[] nums) {
        if(nums.length==0){
            return -1;
        }

        int majo=nums[0];
        int count=1;
        for(int i=1;i<nums.length;i++){
            if(nums[i]==majo){
                count++;
            }
            else{
                count--;
                if(count==0){
                    majo=nums[i];
                    count=1;
                }
            }
        }
        return majo;
    }
}

class Solution

{

    // arr[]: Input Array
    // N : Size of the Array arr[]
    //Function to count inversions in the array.
    static int count=0;
    static long inversionCount(long arr[], long N)
    {
        mergesort(arr,0,N-1);
        return count;
        
    }
    public static void mergesort(long[]arr,long s,long e){
        if(s>e){
            return;
        }
        long mid=s+(e-s)/2;
        mergesort(arr,s,mid);
        mergesort(arr,mid+1,e);
        merge(arr,s,e,mid);
    }
    public static void merge(long[]arr,long s,long e,long mid){
        int i=(int)s;
        int j=(int)mid+1;
        
        while(i<=mid && j<=e){
            if(arr[i]>arr[j]){
                count++;
            }
            i++;
            j++;
        }
    }
}

class Solution

{

    ArrayList<Integer> commonElements(int A[], int B[], int C[], int n1, int n2, int n3) 
    {
        // code here 
        ArrayList<Integer>l=new ArrayList<>();
        
        HashSet<Integer>h1=new HashSet<>();
        HashSet<Integer>h2=new HashSet<>();
        
        for(int i=0;i<n1;i++){
            h1.add(A[i]);
        }
        for(int j=0;j<n2;j++){
            h2.add(B[j]);
        }
        for(int k=0;k<n3;k++){
            if(h1.contains(C[k]) && h2.contains(C[k])){
                if(l.contains(C[k])==false){
                l.add(C[k]);
                }
            }
        }
        
        return l;
    }
}
