package Assignment_Q3;

import java.util.*;

public class Q1 {

    public static void main(String[] args) {
        int nums[]={1,2,3,4,-5};
        int target=10;
        int ans= threeSumClosest(nums,target);
        System.out.println(ans);
    }

    public static  int threeSumClosest(int[] nums, int target) {
        Arrays.sort(nums);
        int cs=Integer.MAX_VALUE;
        int ans=0;
        for(int i=0;i<nums.length-1;i++){
            int l=i+1;
            int h=nums.length-1;

            while(l<h){
                int sum=nums[i]+nums[l]+nums[h];
                if(sum==target){
                    ans=sum;
                    return ans;
                }
                else if(Math.abs(target-sum)<Math.abs(target-cs)){
                    cs=sum;
                }
                else if(sum>target){
                    h--;
                }
                else{
                    l++;
                }
            }

        }
        return cs;
}}

package Assignment_Q3;
import java.util.*;

public class Q2 {

    public static void main(String[] args) {
        int[] nums = {1,0,-1,0,-2,2};
        List<List<Integer>>ans=fourSum(nums,0);
        System.out.println(ans);
    }
    public static List<List<Integer>> fourSum(int[] nums, int target) {
        Arrays.sort(nums);
        List<List<Integer>>ans=new ArrayList<>();
        for(int i=0;i<nums.length-2;i++){
            if(i>0 && nums[i]==nums[i-1]){
                continue;
            }
            for(int j=i+1;j<nums.length-1;j++){
                if(j>i+1 && nums[j]==nums[j-1]){
                    continue;
                }

                int k=j+1;
                int l=nums.length-1;
                while(k<l){
                    long sum=nums[i]+nums[j];
                    sum+=nums[k];
                    sum+=nums[l];

                    if(sum==target){
                        List<Integer>l1=new ArrayList<>();
                        l1.add(nums[i]);
                        l1.add(nums[j]);
                        l1.add(nums[k]);
                        l1.add(nums[l]);
                        ans.add(l1);
                        k++;
                        l--;
                        while(k<l && nums[k]==nums[k-1]){
                            k++;
                        }
                        while(k<l && nums[l]==nums[l+1]){
                            l--;
                        }
                    }
                    else if(sum>target){
                        l--;
                    }else{
                        k++;
                    }
                }
            }
        }
        return ans;
    }
}

package Assignment_Q3;

import java.util.Arrays;

public class Q3 {

    public static void main(String[] args) {
        int[]nums = {1,2,3};
        nextPermutation( nums);
        System.out.println(Arrays.toString(nums));
    }
    public static void nextPermutation(int[] nums) {
        if(nums.length<3){
            reverse(nums,0,nums.length-1);
            return;
        }
        int ind=-1;
        int n=nums.length;

        for(int i=n-2;i>=0;i--){
            if(nums[i]<nums[i+1]){
                ind=i;
                break;
            }
        }
        if(ind==-1){
            reverse(nums,0,nums.length-1);
            return;
        }

        for(int i=n-1;i>=0;i--){
            if(nums[i]>nums[ind]){
                int temp=nums[i];
                nums[i]=nums[ind];
                nums[ind]=temp;

                break;
            }
        }
        reverse(nums,ind+1,nums.length-1);


    }
    public static void reverse(int[]arr,int start,int end){
        while(start<=end){
            int temp=arr[start];
            arr[start]=arr[end];
            arr[end]=temp;
            start++;
            end--;
        }
    }
}

package Assignment_Q3;

public class Q4 {

    public static void main(String[] args) {
        int[] nums = {1,3,5,6};
        System.out.println(searchInsert(nums,5));

    }
    public static int searchInsert(int[] nums, int target) {
        if(target<nums[0]){
            return 0;
        }else if(target>nums[nums.length-1]){
            return nums.length;
        }
        int si=0;
        int ei=nums.length-1;

        int mid=si+(ei-si)/2;

        while(mid<=ei){
            if(nums[mid]==target){
                return mid;
            }else if(target>nums[mid]){
                si=mid+1;
            }else if(target<nums[mid]){
                ei=mid-1;
            }
            mid=si+(ei-si)/2;
        }
        return mid;
    }
}

package Assignment_Q3;

import java.util.*;

public class Q6 {

    public static void main(String[] args) {
        int []nums = {4,1,2,1,2};
        int ans=singleNumber(nums);
        System.out.println(ans);
    }
    public static int singleNumber(int[] nums) {
        HashMap<Integer,Integer> map=new HashMap<>();

        for(int i=0;i<nums.length;i++){
            if(map.containsKey(nums[i])){
                map.put(nums[i],map.get(nums[i])+1);
            }
            else{
                map.put(nums[i],1);
            }
        }

        for(int i=0;i<nums.length;i++){
            if(map.get(nums[i])==1){
                return nums[i];
            }
        }
        return -1;
    }
}

package Assignment_Q3;

import java.util.*;

public class Q7 {

    public static void main(String[] args) {
        int[]nums = {0,1,2,4,5,7};
        List<String>str=summaryRanges(nums);
        System.out.println(str);
    }
    public static List<String> summaryRanges(int[] nums) {
        List<String>l=new ArrayList<>();

        for(int i=0;i<nums.length;i++){
            if(i==nums.length-1){
                String s1=Integer.toString(nums[i]);
                l.add(s1);
                return l;
            }
            int temp=nums[i];
            while(nums[i]+1==nums[i+1]){
                i++;
                if(i==nums.length-1){
                    break;
                }
            }
            int e=nums[i];
            if(temp!=e){
                String s=temp+"->"+e;
                l.add(s);
            }else{
                l.add(Integer.toString(temp));
            }

        }
        return l;
    }

}
